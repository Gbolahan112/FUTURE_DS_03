```python
# Clean column names if needed
df.columns = df.columns.str.strip()
df
```

```python
# Check if there's an index column to drop
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
# Define rating columns (excluding Student ID)
rating_columns = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond',
    'Course recommendation based on relevance'
]
```

```python
# 1: What is the overall average rating across all feedback categories?
average_rating = df[rating_columns].mean().mean()
print(f"The overall average rating across all feedback categories is: {average_rating:.2f}")
```

```python
# Calculate the average rating for each individual feedback category
category_averages = df.drop(columns=['Student ID']).mean()

# Display the results in a readable format
print("Average Ratings by Category:")
for category, avg in category_averages.items():
    print(f"{category}: {avg:.2f}")
```

```python
# Visualize the results with a bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
category_averages.plot(kind='bar')
plt.title('Average Ratings by Feedback Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 2 Highest and Lowest Satisfaction

# Find the highest rated category
highest_category = category_averages.idxmax()
highest_rating = category_averages.max()

# Display the results
print(f"Highest satisfaction: {highest_category} with average rating of {highest_rating:.2f}")
```

```python
# Find the lowest rated category
lowest_category = category_averages.idxmin()
lowest_rating = category_averages.min()
print(f"Lowest satisfaction: {lowest_category} with average rating of {lowest_rating:.2f}")
```

```python
# Visualize with a horizontal bar chart, highlighting the highest and lowest
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
bars = plt.barh(category_averages.index, category_averages.values)

# Highlight the highest and lowest bars
colors = ['lightblue'] * len(category_averages)
colors[category_averages.values.argmax()] = 'green'
colors[category_averages.values.argmin()] = 'red'

for i, bar in enumerate(bars):
    bar.set_color(colors[i])

plt.xlabel('Average Rating')
plt.title('Feedback Categories by Satisfaction Level')
plt.tight_layout()
plt.show()
```

```python
#3 SUBJECT KNOWLEDGE RATING
subject_knowledge=df["Well versed with the subject"]
print(f"Subject Knowledge Rating Analysis:")
```

```python
# Calculate basic statistics
avg_rating = subject_knowledge.mean()
median_rating = subject_knowledge.median()
min_rating = subject_knowledge.min()
max_rating = subject_knowledge.max()
```

```python
# Count the frequency of each rating
rating_counts = subject_knowledge.value_counts().sort_index()
```

```python
# Display the results
print(f"Subject Knowledge Rating Analysis:")
print(f"Average rating: {avg_rating:.2f}")
print(f"Median rating: {median_rating}")
print(f"Range: {min_rating} to {max_rating}")
print("\nRating distribution:")
for rating, count in rating_counts.items():
    percentage = (count / len(subject_knowledge)) * 100
    print(f"Rating {rating}: {count} students ({percentage:.1f}%)")
```

```python
# Visualize the distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(subject_knowledge, bins=range(min_rating, max_rating+2), edgecolor='black', alpha=0.7)
plt.title('Distribution of Subject Knowledge Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Students')
plt.xticks(range(min_rating, max_rating+1))
plt.axvline(avg_rating, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {avg_rating:.2f}')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# 4 Correlation Between Categories
# Are there certain feedback categories correlated with each other?
# Calculate correlation matrix
corr_matrix = df.drop(columns=['Student ID']).corr()
```

```python
# Visualize the correlations
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Between Feedback Categories')
plt.tight_layout()
plt.show()
```

```python
# Find the strongest positive correlation (excluding self-correlations)
corr_no_self = corr_matrix.where(~np.eye(corr_matrix.shape[0], dtype=bool))
max_corr = corr_no_self.max().max()
max_corr_pair = np.where(corr_no_self == max_corr)
print(f"Strongest positive correlation ({max_corr:.2f}) is between:")
print(f"- {corr_matrix.index[max_corr_pair[0][0]]}")
print(f"- {corr_matrix.columns[max_corr_pair[1][0]]}")
```

```python
# 5 Overall Satisfaction Distribution
# What's the distribution of overall satisfaction across all categories?
# Calculate average rating per student
student_avg_ratings = df.drop(columns=['Student ID']).mean(axis=1)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(student_avg_ratings, kde=True, bins=20)
plt.title('Distribution of Overall Student Satisfaction')
plt.xlabel('Average Rating Across All Categories')
plt.ylabel('Number of Students')
plt.axvline(student_avg_ratings.mean(), color='red', linestyle='dashed', 
            label=f'Mean: {student_avg_ratings.mean():.2f}')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Calculate percentiles
percentiles = [10, 25, 50, 75, 90]
for p in percentiles:
    print(f"{p}th percentile: {np.percentile(student_avg_ratings, p):.2f}")
```

```python
# 6  Identify Potential Outliers
# Are there any outlier ratings that might need attention?
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.drop(columns=['Student ID']))
plt.title('Boxplot of Ratings by Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Find potential outlier students (those with consistently low or high ratings)
df['avg_rating'] = df.drop(columns=['Student ID']).mean(axis=1)
outliers = df[(df['avg_rating'] < df['avg_rating'].quantile(0.1)) | 
              (df['avg_rating'] > df['avg_rating'].quantile(0.9))]

print(f"Potential outlier students (very high or very low ratings):")
print(outliers[['Student ID', 'avg_rating']])
```

```python
# 7 Gap Analysis
# What's the gap between expectations and delivery?
# Let's compare "Course recommendation based on relevance" with actual delivery metrics

relevance = df['Course recommendation based on relevance']
delivery = df[['Explains concepts in an understandable way', 
               'Use of presentations', 
               'Structuring of the course']].mean(axis=1)

plt.figure(figsize=(10, 6))
plt.scatter(relevance, delivery, alpha=0.5)
plt.plot([min(relevance), max(relevance)], [min(relevance), max(relevance)], 
         'r--', label='Perfect alignment')
plt.xlabel('Course Recommendation Based on Relevance')
plt.ylabel('Average Delivery Quality')
plt.title('Expectation vs. Delivery Gap Analysis')
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# Calculate the average gap
gap = delivery - relevance
print(f"Average gap between delivery and expectations: {gap.mean():.2f}")
print(f"Positive value means delivery exceeds expectations")
print(f"Negative value means delivery falls short of expectations")
```

```python
# 8 Identify Areas for Improvement
# Which categories have the most room for improvement?
category_averages = df.drop(columns=['Student ID']).mean()
room_for_improvement = 5 - category_averages  # Assuming 5 is the max rating

plt.figure(figsize=(10, 6))
room_for_improvement.sort_values().plot(kind='barh')
plt.title('Room for Improvement by Category')
plt.xlabel('Gap to Perfect Score (5)')
plt.tight_layout()
plt.show()

print("Top 3 areas with most room for improvement:")
for category, gap in room_for_improvement.sort_values(ascending=False).head(3).items():
    print(f"- {category}: {gap:.2f} points from perfect score")
```
